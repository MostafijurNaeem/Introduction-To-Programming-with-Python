```python
import ipywidgets as widgets
from IPython.display import display

# Initialize the game state
board = [[" " for _ in range(3)] for _ in range(3)]
current_player = "X"

# Function to check for a win
def check_win():
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != " ":
            return True
        if board[0][i] == board[1][i] == board[2][i] != " ":
            return True
    if board[0][0] == board[1][1] == board[2][2] != " ":
        return True
    if board[0][2] == board[1][1] == board[2][0] != " ":
        return True
    return False

# Function to check if the board is full
def is_full():
    return all(cell != " " for row in board for cell in row)

# Button click handler
def on_click(change, row, col):
    global current_player
    if board[row][col] == " " and not check_win():
        board[row][col] = current_player
        buttons[row][col].description = current_player
        buttons[row][col].style.button_color = "darkred" if current_player == "X" else "green"
        buttons[row][col].disabled = True

        if check_win():
            print(f"Player {current_player} wins!")
            disable_all_buttons()
        elif is_full():
            print("It's a draw!")
        else:
            current_player = "O" if current_player == "X" else "X"

# Disable all buttons (when game ends)
def disable_all_buttons():
    for row in buttons:
        for btn in row:
            btn.disabled = True

# Create the grid of buttons
buttons = [[widgets.Button(description=" ", layout=widgets.Layout(width="60px", height="60px"), style={"button_color": "white"}) for _ in range(3)] for _ in range(3)]
grid = widgets.GridBox(children=[btn for row in buttons for btn in row],
                       layout=widgets.Layout(grid_template_columns="60px 60px 60px", grid_gap="5px"))

# Attach click handlers to buttons
for i in range(3):
    for j in range(3):
        buttons[i][j].on_click(lambda change, row=i, col=j: on_click(change, row, col))

# Display the game grid
display(grid)
print(f"Player {current_player}'s turn!")
```
